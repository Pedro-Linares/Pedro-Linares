```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
xx=np.linspace(-5,5,100)
yy=np.linspace(-5,5,100)
x,y=np.meshgrid(xx,yy)
f=(1-x)**2 + 2*(3-y)**2
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(x,y,f)
from sympy import *
x,y,f=symbols ('x y f')
f=x**2+y**2
fx=diff(f,x)
fy=diff(f,y)
fxx=diff(f,x,2)
fyy=diff(f,y,2)
fxy=diff(fx,y)
fyx=diff(fy,x)
px=solve(Eq(fx,0))
py=solve(Eq(fy,0))
fxxp=fxx.subs({x:px[0],y:py[0]})
fyyp=fyy.subs({x:px[0],y:py[0]})
fxyp=fxy.subs({x:px[0],y:py[0]})
fyxp=fyx.subs({x:px[0],y:py[0]})
d=fxxp*fyyp-fxyp*fxyp
print('solução: (%.2f, %.2f)' % (px[0],py[0]))
print('Determinante:', d)
print('Derivada Segunda em relação a "x":', fxxp)
```

**Tema 05 Sistemas lineares com números complexos**

```python
# sistema linear complexo:
# (2+j)a + (5-j)b = 3+8j
# (5-j)a + (7+4j)b = 2+5j
import numpy as np
A=np.array([[complex(2,1),complex(5,-1)],[complex(5,-1), complex(7,4)]])
b=np.array([[complex(3,8)],[complex(2,5)]])
np.linalg.solve(A,b)
```

```python
# sistema linear complexo:
# (3+2j)a + (-2-6j)b = 70<0º (70=raio(r))
# (-2-6j)a + (10+j)b = 110<30º (110=raio(r))
import numpy as np
A=np.array([[complex(3,2),complex(-2,-6)],[complex(-2,-6), complex(10,1)]])
x1=np.cos(np.deg2rad(0))
y1=np.sin(np.deg2rad(0))
x2=np.cos(np.deg2rad(30))
y2=np.sin(np.deg2rad(30))
b=np.array([[70*complex(x1,y1)],[110*complex(x2,y2)]])
np.linalg.solve(A,b)
```
